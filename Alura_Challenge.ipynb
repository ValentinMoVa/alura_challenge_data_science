```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

```python
ing_tienda1 = tienda1['Precio'].sum()
ing_tienda2 = tienda2['Precio'].sum()
ing_tienda3 = tienda3['Precio'].sum()
ing_tienda4 = tienda4['Precio'].sum()
print(f'Los ingresos de la tienda 1 son iguales a {ing_tienda1}')
print(f'Los ingresos de la tienda 2 son iguales a {ing_tienda2}')
print(f'Los ingresos de la tienda 3 son iguales a {ing_tienda3}')
print(f'Los ingresos de la tienda 4 son iguales a {ing_tienda4}')
```

```python
contador_categorias1 = tienda1['Categoría del Producto'].value_counts()
print(contador_categorias1)
```

```python
contador_categorias2 = tienda2['Categoría del Producto'].value_counts()
print(contador_categorias2)
```

```python
contador_categorias3 = tienda3['Categoría del Producto'].value_counts()
print(contador_categorias3)
```

```python
contador_categorias4 = tienda4['Categoría del Producto'].value_counts()
print(contador_categorias4)
```

```python
calificacion_tienda1 = tienda1['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()

print(f'La calificación promedio de la tienda 1 es de {calificacion_tienda1:.2f}')
print(f'La calificación promedio de la tienda 2 es de {calificacion_tienda2:.2f}')
print(f'La calificación promedio de la tienda 3 es de {calificacion_tienda3:.2f}')
print(f'La calificación promedio de la tienda 4 es de {calificacion_tienda4:.2f}')
```

```python
prom_envio1=tienda1['Costo de envío'].mean()
prom_envio2=tienda2['Costo de envío'].mean()
prom_envio3=tienda3['Costo de envío'].mean()
prom_envio4=tienda4['Costo de envío'].mean()
print(f'El promedio de costo de envío de la tienda 1 es de {prom_envio1:.2f}')
print(f'El promedio de costo de envío de la tienda 2 es de {prom_envio2:.2f}')
print(f'El promedio de costo de envío de la tienda 3 es de {prom_envio3:.2f}')
print(f'El promedio de costo de envío de la tienda 4 es de {prom_envio4:.2f}')
```

```python
union_tiendas=pd.concat([tienda1,tienda2,tienda3,tienda4])
vendedor=union_tiendas['Vendedor'].value_counts()

etiquetas = vendedor.index.tolist()
porcentajes = vendedor.values.tolist()

plt.bar(etiquetas, porcentajes)
plt.title('Vendedores y su numero de vantas')
plt.xlabel('Vendedor')
plt.ylabel('Numero de Ventas')
plt.xticks(rotation=90)  # Optional: to make labels more readable
plt.tight_layout()       # Optional: adjusts layout to prevent clipping
plt.show()
```
